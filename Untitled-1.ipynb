```python
import numpy as np
import faiss

# 1. إنشاء بيانات عشوائية (مثلاً 100 vector كل واحد طوله 5)
d = 5  # عدد الأبعاد
nb = 100  # عدد الـ vectors في قاعدة البيانات
np.random.seed(42)
data = np.random.random((nb, d)).astype('float32')

# 2. إنشاء index باستخدام L2 distance (المسافة الإقليدية)
index = faiss.IndexFlatL2(d)  # flat index = لا ضغط ولا تعقيد، مباشر
index.add(data)  # أضف البيانات إلى الفهرس

print("Number of vectors in index:", index.ntotal)

# 3. ابحث عن أقرب 3 جيران لنقطة جديدة
query = np.random.random((1, d)).astype('float32')
distances, indices = index.search(query, k=3)

print("Query Vector:", query)
print("Nearest Neighbors' Indices:", indices)
print("Distances:", distances)
```
